```python
# Features
X = df.drop("Churn Label", axis=1)

# Target
y = df["Churn Label"]

print(X.shape)
print(y.shape)
```

```python
print(X.dtypes)
```

```python
X = pd.get_dummies(
    X,
    drop_first=True,
    dtype=int
)
```

```python
print(X.head())
```

```python
print(X.shape)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.20,
    random_state=42,
    stratify=y
)
```

```python
print("Training Set :", X_train.shape)
print("Testing Set  :", X_test.shape)

print("\nTraining Labels")
print(y_train.value_counts(normalize=True))

print("\nTesting Labels")
print(y_test.value_counts(normalize=True))
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report
)

results = []

def evaluate_model(model, X_test, y_test):

    # Predictions
    y_pred = model.predict(X_test)
    y_prob = model.predict_proba(X_test)[:, 1]

    # Metrics
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred, pos_label="Yes")
    recall = recall_score(y_test, y_pred, pos_label="Yes")
    f1 = f1_score(y_test, y_pred, pos_label="Yes")
    roc_auc = roc_auc_score((y_test=="Yes"), y_prob)

    print(classification_report(y_test, y_pred))

    cm = confusion_matrix(y_test, y_pred)

    plt.figure(figsize=(5,4))

    sns.heatmap(
        cm,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=["No","Yes"],
        yticklabels=["No","Yes"]
    )

    plt.xlabel("Predicted")
    plt.ylabel("Actual")
    plt.show()

    return accuracy, precision, recall, f1, roc_auc
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier(
    random_state=42
)
```

```python
dt_model.fit(X_train, y_train)
```

```python
accuracy, precision, recall, f1, roc_auc = evaluate_model(
    dt_model,
    X_test,
    y_test
)
```

```python
results.append({
    "Model": "Decision Tree",
    "Accuracy": accuracy,
    "Precision": precision,
    "Recall": recall,
    "F1 Score": f1,
    "ROC AUC": roc_auc
})
```

```python
feature_importance = pd.DataFrame({
    "Feature": X.columns,
    "Importance": dt_model.feature_importances_
})

feature_importance = feature_importance.sort_values(
    by="Importance",
    ascending=False
)

feature_importance.head(10)

plt.figure(figsize=(10,6))

sns.barplot(
    data=feature_importance.head(10),
    x="Importance",
    y="Feature"
)

plt.title("Top 10 Important Features (Random Forest)")

plt.show()
```

```python
# Convert target labels
y_train_xgb = y_train.map({"No": 0, "Yes": 1})
y_test_xgb = y_test.map({"No": 0, "Yes": 1})
```

```python
print(y_train_xgb.unique())
print(y_test_xgb.unique())
```

```python
xgb_model = XGBClassifier(
    n_estimators=100,
    learning_rate=0.1,
    max_depth=6,
    random_state=42,
    eval_metric="logloss"
)
```

```python
xgb_model.fit(
    X_train,
    y_train_xgb
)
```

```python
y_pred = xgb_model.predict(X_test)
y_prob = xgb_model.predict_proba(X_test)[:, 1]
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score,
    roc_auc_score,
    confusion_matrix,
    classification_report
)
```

```python
print(classification_report(y_test_xgb, y_pred))
accuracy = accuracy_score(y_test_xgb, y_pred)

precision = precision_score(y_test_xgb, y_pred)

recall = recall_score(y_test_xgb, y_pred)

f1 = f1_score(y_test_xgb, y_pred)

roc_auc = roc_auc_score(y_test_xgb, y_prob)
```

```python
print("XGBoost Performance")
print("-" * 40)

print(f"Accuracy : {accuracy:.4f}")
print(f"Precision: {precision:.4f}")
print(f"Recall   : {recall:.4f}")
print(f"F1 Score : {f1:.4f}")
print(f"ROC AUC  : {roc_auc:.4f}")
```

```python
cm = confusion_matrix(y_test_xgb, y_pred)

plt.figure(figsize=(5,4))

sns.heatmap(
    cm,
    annot=True,
    fmt="d",
    cmap="Blues",
    xticklabels=["No", "Yes"],
    yticklabels=["No", "Yes"]
)

plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("XGBoost Confusion Matrix")

plt.show()
```
